# Exercise 2: Visualization

## Visualization Types
We'll cover both **univariate** (single variable) and **multivariate** (multiple variables) visualizations.

**Question 1**: Create a univariate bar plot showing the count of products in each category.

```python
# Your code here
nb_products = df.groupby('Category')['Quantity'].sum()
plt.figure(figsize=(10, 6))
plt.bar(nb_products.index, nb_products.values, color='steelblue', edgecolor='black')
plt.title('Total Quantity Sold per Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Question 2**: Create a univariate histogram showing the distribution of prices.


```python
# Your code here
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=30, alpha=0.7, color='steelblue', edgecolor='black')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Question 3**: Create a univariate box plot showing the distribution of quantities sold.


```python
# Your code here
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.boxplot(df['Quantity'], vert=False)
plt.title('Distribution of Quantities Sold')
plt.xlabel('Quantity')
plt.grid(axis='x', alpha=0.3)
plt.subplot(1, 2, 2)
plt.hist(df['Quantity'], bins=9, alpha=0.7, color='steelblue', edgecolor='black')
plt.title('Distribution of Quantities Sold')
plt.xlabel('Quantity')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Question 4**: Create a multivariate scatter plot showing the relationship between price and quantity.


```python
# Your code here
plt.figure(figsize=(10, 6))
plt.scatter(df['Price'], df['Quantity'], alpha=0.5, color='steelblue')
plt.title('Price vs Quantity')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.grid(alpha=0.3)
plt.tight_layout()
```

**Question 5**: Create a multivariate histogram showing the distribution of prices grouped by category.


```python
# Your code here
plt.figure(figsize=(10, 6))
colors = ['yellow', 'red', 'blue', 'green'] 
for category, color in zip(df['Category'].unique(), colors):
    plt.hist(df[df['Category'] == category]['Price'], 
             bins=30, 
             alpha=0.7, 
             color=color, 
             edgecolor='black',
             label=category)
plt.legend(title='Category')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
```

**Question 6**: Create a multivariate box plot showing the distribution of prices by region.



```python
# Your code here
plt.figure(figsize=(10, 6))
plt.boxplot([df['Price'][df['Region'] == region] for region in df['Region'].unique()], 
            vert=False)
plt.title('Distribution of Prices by Region')
plt.xlabel('Price')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
```

**Question 7**: Create a multivariate colored scatter plot showing price vs quantity, with colors representing different categories.



```python
# Your code here
plt.figure(figsize=(10, 6))
plt.scatter(df['Price'], df['Quantity'], alpha=0.5, 
            c = df['Category'].map({'Electronics': 'red', 'Clothing': 'blue', 'Books': 'green', 'Home': 'yellow'}))
plt.title('Price vs Quantity Colored by Category')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
```

```python
# Load the dataset
df = pd.read_csv('/Users/gaspardvilla/src/Nomades_ADA/data/week_2/waves.csv', index_col=0, parse_dates=True)

# Display basic information about the dataset
df.info()
df.describe()
```

**Question 2**: Create comprehensive visualizations of your dataset using all the plot types from Exercise 2. Some insights are hidden into this dataset, try to find them!


```python
# Set up the plotting style
plt.style.use('default')
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Surfing Wave Data Analysis - Key Relationships', fontsize=16, fontweight='bold')

# 1. Wind speed distribution (histogram) - show bimodal pattern
axes[0, 0].hist(df['wind_speed'], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Wind Speed Distribution', fontweight='bold')
axes[0, 0].set_xlabel('Wind Speed (knots)')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(True, alpha=0.3)

# 2. Wave height distribution (histogram)
axes[0, 1].hist(df['wave_height'], bins=30, alpha=0.7, color='lightcoral', edgecolor='black')
axes[0, 1].set_title('Wave Height Distribution', fontweight='bold')
axes[0, 1].set_xlabel('Wave Height (meters)')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].grid(True, alpha=0.3)

# 3. Wind speed vs Hour of day (box plot)
hour_data = [df[df.index.hour == hour]['wind_speed'].values for hour in range(24)]
axes[0, 2].boxplot(hour_data, positions=range(24))
axes[0, 2].set_title('Wind Speed by Hour of Day', fontweight='bold')
axes[0, 2].set_xlabel('Hour of Day')
axes[0, 2].set_ylabel('Wind Speed (knots)')
axes[0, 2].set_xticks(range(0, 24, 1))
axes[0, 2].grid(True, alpha=0.3)

# 4. Wave height vs Day of week (box plot)
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_data = [df[df.index.weekday == day]['wave_height'].values for day in range(7)]
axes[1, 0].boxplot(day_data, labels=day_names)
axes[1, 0].set_title('Wave Height by Day of Week', fontweight='bold')
axes[1, 0].set_xlabel('Day of Week')
axes[1, 0].set_ylabel('Wave Height (meters)')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].grid(True, alpha=0.3)

# 5. Wave height vs Month (box plot)
month_data = [df[df.index.month == month]['wave_height'].values for month in [8, 9]]
axes[1, 1].boxplot(month_data, labels=['August', 'September'])
axes[1, 1].set_title('Wave Height by Month', fontweight='bold')
axes[1, 1].set_xlabel('Month')
axes[1, 1].set_ylabel('Wave Height (meters)')
axes[1, 1].grid(True, alpha=0.3)

# 6. Additional insight: Wind speed vs Wave height scatter plot
axes[1, 2].scatter(df['wind_speed'], df['wave_height'], alpha=0.5, color='green')
axes[1, 2].set_title('Wind Speed vs Wave Height', fontweight='bold')
axes[1, 2].set_xlabel('Wind Speed (knots)')
axes[1, 2].set_ylabel('Wave Height (meters)')
axes[1, 2].grid(True, alpha=0.3)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
