```python
plt.figure(figsize=(10, 6))

# Create color mapping for basement (green for yes, red for no)
basement_colors = ['green' if x == 'yes' else 'red' for x in df['basement']]

# Create marker mapping for parking (o for parking, x for no parking)
parking_markers = ['o' if x > 0 else 'x' for x in df['parking']]

# Plot each point with its specific color and marker
for i, (price, area, color, marker) in enumerate(zip(df['price'], df['area'], basement_colors, parking_markers)):
    plt.scatter(price, area, c=color, marker=marker, alpha=0.6, s=50)

plt.xlabel('Price')
plt.ylabel('Age')
plt.title('Housing Prices in San Francisco')
plt.show()
```

```python
import scipy.stats as stats

# Generate a plot from binomial distribution and plot the PDF and CDF of the 
# given data produced
rv = stats.binom(n=10, p=0.5)
colors = ['darkred', 'darkblue', 'darkgreen', 'orange']

# Plot for size in [10, 100, 1000, 10000]
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
for color, size, ax in zip(colors, [10, 100, 1000, 10000], axs.flatten()):
    data = rv.rvs(size = size)
    unique, counts = np.unique(data, return_counts=True)

    # Make sure we go from 0 to 10
    for i in range(11):
        if i not in unique:
            unique = np.append(unique, i)
            counts = np.append(counts, 0)

    # Plot the PDF
    ax.bar(unique, counts / size, label='PDF', color=color)
    ax.set_xlabel('Number of successes')
    ax.set_title(f'Size = {size}')

plt.tight_layout()
plt.show()

```

```python
import scipy.stats as stats

# Generate a plot from binomial distribution and plot the PDF and CDF of the 
# given data produced
rv = stats.binom(n=10, p=0.5)

# Plot for size in [10, 100, 1000, 10000]
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
for color, size, ax in zip(colors, [10, 100, 1000, 10000], axs.flatten()):
    data = rv.rvs(size = size)
    unique, counts = np.unique(data, return_counts=True)
    cumulative = np.cumsum(counts)

    # Make sure we go from 0 to 10
    for i in range(11):
        if i not in unique:
            unique = np.append(unique, i)
            counts = np.append(counts, 0)
            cumulative = np.append(cumulative, 0)

    # Plot the PDF
    ax.bar(unique, cumulative, label='CDF', color=color)
    ax.set_xlabel('Number of successes')
    ax.set_title(f'Size = {size}')

plt.tight_layout()
plt.show()

```
